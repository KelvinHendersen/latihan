```python
import pandas as pd
import os
```

```python
TIPE_PLOT_BAR = 'bar'
TIPE_PLOT_BAR_HORIZONTAL = 'barh'
TIPE_PLOT_LINE = 'line'
```

```python
current_path = os.getcwd() # dapetin current path
filename = 'Data Faskes BPJS 2019.csv'
data_csv = pd.read_csv(os.path.join(current_path,filename ))
```

```python
data_csv.head(5)
```

```python
print("Provinsi: ", data_csv['Provinsi'].unique().tolist())
```

```python
data_csv['Provinsi'].value_counts().plot.bar(title="Jumlah Faskes di Setiap Provinsi")
```

```python
data_csv['TipeFaskes'].value_counts().plot.bar(title="Jumlah Faskes di Indonesia Berdasarkan Tipe")
```

```python
def display_by_certain_column(data_csv, nama_kolom_filter, nama_filter, nama_show_kolom, title, tipe_plot):
  df = data_csv.loc[data_csv[nama_kolom_filter] == nama_filter]
  if tipe_plot == TIPE_PLOT_BAR:
    df[nama_show_kolom].value_counts().plot.bar(title=title)
    return
  if tipe_plot == TIPE_PLOT_BAR_HORIZONTAL:
    df[nama_show_kolom].value_counts().plot.barh(title=title)
    return
  if tipe_plot == TIPE_PLOT_LINE:
    df[nama_show_kolom].value_counts().plot.line(title=title)
    return
  
  return
```

```python
display_by_certain_column(data_csv, "Provinsi", "Jawa Tengah", "TipeFaskes", "Jumlah Tipe Faskes di Jawa Tengah", TIPE_PLOT_BAR_HORIZONTAL)
```
